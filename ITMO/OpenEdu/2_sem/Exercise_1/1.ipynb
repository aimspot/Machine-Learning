```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA

df = pd.read_csv("data/34_16.csv", header=None)

```

```python
df.head()
```

```python
arr = df.to_numpy()
arr
```

```python
center_function = lambda x: x - x.mean()
F = center_function(arr)
F
```

```python
pca = PCA(n_components=23)
principalComponents = pca.fit_transform(F)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])



variance = pca.explained_variance_ratio_[0]+pca.explained_variance_ratio_[1]
# Объясненная дисперсия
print(round(variance, 4))
# # print(principalDf)
m = principalDf.to_numpy()

vec1 = m[:,0]
vec2 = m[:,1]
     
```

```python
principalDf
```

```python


vec2
```

```python

import matplotlib.pyplot as plt

fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
colors = ['r', 'g', 'b']


ax.scatter(vec1,vec2
              #  , c = colors
               , s = 50)
# ax.legend(targets)
ax.grid()
```
